```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("B:\\prog\\p3\\charting-w-stats-ShotDirection.csv")
# calculer le nombre total de balles reçues dans chaque zone
zone_totals = df[['crosscourt', 'down_middle', 'down_the_line', 'inside_out', 'inside_in']].sum()

# trouver la zone ayant reçu le plus de balles
zone_max = zone_totals.idxmax()

# créer un diagramme en bâtons des totaux pour chaque zone
zone_totals.plot(kind='bar')

# dessiner un terrain de tennis avec la zone ayant reçu le plus de balles en surbrillance
if zone_max == 'crosscourt':
    plt.axhspan(0, 25, facecolor='red', alpha=0.5)
elif zone_max == 'down_middle':
    plt.axhspan(25, 50, facecolor='red', alpha=0.5)
elif zone_max == 'down_the_line':
    plt.axhspan(50, 75, facecolor='red', alpha=0.5)
elif zone_max == 'inside_out':
    plt.axhspan(75, 100, facecolor='red', alpha=0.5)
elif zone_max == 'inside_in':
    plt.axhspan(0, 100, facecolor='red', alpha=0.5)
    
# ajouter des étiquettes pour les zones
plt.xticks(range(5), ['Crosscourt', 'Down Middle', 'Down the Line', 'Inside Out', 'Inside In'])

# ajouter une étiquette pour l'axe des y
plt.ylabel('Nombre de balles reçues')

# afficher le graphique
plt.show()
```
